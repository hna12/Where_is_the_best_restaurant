```python
#맛집list 가져오기
def best_restaurant(location):

  poing = {
    'name': [],
    'stars': [],
    'address' : [],
    'food type' : [],
    'review count' : []
  }
  import requests
  from bs4 import BeautifulSoup
  import json
  import pandas as pd
  from tqdm.notebook import tqdm

  for i in tqdm(range(1,20)):
    url = f'https://poing-gw.mypoing.com/place?offset={i}&limit=20&is_schedule=1&area={location}&sort=recommended&is_random=1'
    headers = {
      'Poing-Authorization-Key': 'efd955333aff804e494bcaaa80ecf786',
      'Poing-Authorization-Name': 'poing-3.1.0',
      'Referer': 'https://m.poing.co.kr/',
      'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.84 Safari/537.36'
    }
    resp = requests.get(url, headers = headers)
    dic_resp = json.loads(resp.text) #offset부분
    datas = dic_resp['data']
    for data in datas:
      poing['name'].append(data['name'])
      poing['address'].append(data['address'])
      poing['food type'].append(data['food_types'][0]['title'])
      poing['stars'].append(data['stars'])
      poing['review count'].append(data['review_count'])
  
  po = pd.DataFrame(poing)
  return po
```

```python
def poing():
  print(location)
  id = input('원하는 위치의 id를 써주세요')
  return best_restaurant(id)
```
