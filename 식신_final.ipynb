```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
##https://www.siksinhot.com/search?keywords=%EC%84%B1%EC%88%98
```

```python
data = {
    '이름' : [],
    '평점' : [],
    '주소' : [],
    '카테고리' : [],
    '조회수' : [],
    # '대표메뉴' : []
}
```

```python
for i in range(1,33):
  url = f'https://api.siksinhot.com/v1/hp/search/place?keywords=%EC%84%B1%EC%88%98&idx={12*i}&limit=12&order=interest'
  headers={
    'referer': 'https://www.siksinhot.com/',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.60 Safari/537.36 Edg/100.0.1185.29',
    'siksinoauth': 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJ1aWQiOjAsImlhdCI6MTY0OTUwOTMyMiwiZXhwIjoxNjQ5NTk1NzIyLCJpc3MiOiJzaWtzaW4ifQ.rxVffd5Bd3T13TyfUJdRtMLR1lLCGety96Xb-urAfhc'
  }
  resp = requests.get(url, headers=headers)
  temp = json.loads(resp.text)
  restaurants = temp['data']['list']
  
  for res in restaurants:
    data['이름'].append(res['pname'])
    data['평점'].append(res['score'])
    data['주소'].append(res['addr2'])
    data['카테고리'].append(res['hpSchCateNm'])
    data['조회수'].append(res['viewCnt'])
test = pd.DataFrame(data)
test
```

```python
df = pd.DataFrame(test)
```
